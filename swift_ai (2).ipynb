```python
from PyPDF2 import PdfReader
import re
import os

# 1️⃣ Read PDF
pdf_path = "usvisa.pdf"  # your PDF file
reader = PdfReader(pdf_path)
text = ""

for page in reader.pages:
    page_text = page.extract_text()
    if page_text:
        text += page_text + " "

# 2️⃣ Preprocessing
# Remove unwanted special characters (keep letters, numbers, comma, period, dash, space)
text = re.sub(r'[^a-zA-Z0-9., -]+', ' ', text)
# Replace multiple spaces with single space
text = re.sub(r'\s+', ' ', text).strip()

print("====== Preprocessed Text ======")
print(text[:1000])  # print first 1000 characters for checking
print("\nTotal characters in text:", len(text))

# 3️⃣ Chunking
chunk_size = 1000  # characters per chunk
chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]

# 4️⃣ Create folder to save chunks (optional)
os.makedirs("chunks", exist_ok=True)

# 5️⃣ Print chunks with character count
for idx, chunk in enumerate(chunks):
    print(f"\n====== Chunk {idx+1} (Characters: {len(chunk)}) ======")
    print(chunk)
    # Optional: Save each chunk as a text file
    with open(f"chunks/chunk_{idx+1}.txt", "w", encoding="utf-8") as f:
        f.write(chunk)
```

```python
from sentence_transformers import SentenceTransformer

# Load the model
model = SentenceTransformer('all-MiniLM-L6-v2', device='cpu')  # CPU version

# Generate embeddings for each chunk
embeddings = [model.encode(chunk) for chunk in chunks]

# Print embeddings with length
for idx, emb in enumerate(embeddings):
    print(f"\n====== Chunk {idx+1} Embedding ======")
    print(f"Characters in chunk: {len(chunks[idx])}")
    print(f"Embedding length: {len(emb)}")
    print(f"Embedding snippet (first 10 values): {emb[:10]}")
```

```python
import faiss
import numpy as np

# Convert embeddings to numpy array
embedding_array = np.array(embeddings, dtype='float32')

# Get dimension
dim = embedding_array.shape[1]

# Create FAISS index
index = faiss.IndexFlatL2(dim)
index.add(embedding_array)

# Print index info
print("===== FAISS Index Info =====")
print(f"Number of vectors in index: {index.ntotal}")
print(f"Dimension of each vector: {dim}\n")

# Print all chunks fully
print("===== Chunks (Full Text) =====")
for i, chunk in enumerate(chunks):
    print(f"\nChunk {i+1}:\n{chunk}\n{'-'*80}")

# Sample search for each chunk with full text
print("\n===== FAISS Search for Each Chunk =====")
k = 3
for i, query_vec in enumerate(embedding_array):
    query_vec = np.expand_dims(query_vec, axis=0)
    distances, indices = index.search(query_vec, k)
    print(f"\nQuery Chunk {i+1}:\n{chunks[i]}\n")
    for j, idx in enumerate(indices[0]):
        print(f"Neighbor {j+1}: Chunk {idx+1} | Distance: {distances[0][j]:.4f}")
        print(f"{chunks[idx]}\n{'-'*80}")


```
